```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the .csv file into a DataFrame
file_path = 'creditcard_2023.csv'
df = pd.read_csv(file_path)

```

### Profile the Data
The first step in any analytics project is to explore the data. For each column in our dataset, we should examine for any outliers, unusual data distributions, or missing values. I will do that below.

```python
#Profile the data
def plot_histograms(df):
    for column in df.columns:
        plt.figure(figsize=(10, 6))
        if pd.api.types.is_numeric_dtype(df[column]):
            df[column].plot(kind='hist', bins=20, title=column)
        else:
            sns.countplot(y=column, data=df)
            plt.title(column)
        plt.show()

plot_histograms(df)
```

```python
#Heatmap of missing values
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()
```

It looks like we have a nice & clean dataset with no missing values and a bell curve on most categorical variables. The 'Class' column shoudl be categorical, so I will make that change now.
